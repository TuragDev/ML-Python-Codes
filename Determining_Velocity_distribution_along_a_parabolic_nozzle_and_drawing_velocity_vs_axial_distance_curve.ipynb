```python
#Velocty distribution along a parabolic nozzle for 100% opening is determined and axial distance vs. flow velocity curve is drawn
#Velocity at a certain point can be known from the manometric height of that point
#Applying Bernoulli's equation, the formula of velocity at the certain point had been derived. Velocity = sqrt((2*g*hw)/spgra)
#Here, g = standard gravity (Considered it as 9.81), hw = Manometric height (Measured by experiment), spgra = specific gravity of air (considered 0.00121)
```

```python
import math
import matplotlib.pyplot as plt
spgra = 0.00121
g = 9.81
```

```python
def parabolic_nozzle(distances,hw_values):
  velocities = [math.sqrt((2*g*hw*0.001)/spgra) for hw in hw_values] #(0.001 is multiplied with hw to convert the value from millimiter to meter)
  plt.plot(distances,velocities,marker='o')
  plt.xlabel('Distance (mm)')
  plt.ylabel('Velocities (m/s)')
  plt.title('Velocity vs. Distance')
  plt.grid(True)
  plt.show()
```

```python
distances=[]
hw_values = []
while True:
   distance = input("Enter distance (in mm) or enter 'done to finish: " )
   if distance == 'done':
      break
   hw = float (input("Enter value of manometric height (in mm) at the distance: "))
   distances.append(distance)
   hw_values.append(hw)
parabolic_nozzle(distances, hw_values)
```
